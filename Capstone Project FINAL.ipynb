### Part II. Analysing Toronto data set

```python
source = requests.get("https://web.archive.org/web/20190825081918/https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
soup = BeautifulSoup(source, 'lxml')

table = soup.find("table")
table_rows = table.tbody.find_all("tr")

res = []
for tr in table_rows:
    td = tr.find_all("td")
    row = [tr.text for tr in td]
    
    if row != [] and row[1] != "Not assigned":
        if "Not assigned" in row[2]: 
            row[2] = row[1]
        res.append(row)

T_df = pd.DataFrame(res, columns = ["PostalCode", "Borough", "Neighborhood"])
T_df.head()
```

```python
T_df["Neighborhood"] = T_df["Neighborhood" ].str.replace("\n","")
T_df["Borough"] = T_df["Borough" ].str.replace("\n","")
T_df["PostalCode"] = T_df["PostalCode" ].str.replace("\n","")
T_df = T_df.groupby(["PostalCode", "Borough"])["Neighborhood"].apply(", ".join).reset_index()
print("Shape: ", T_df.shape)
T_df.head()
```
